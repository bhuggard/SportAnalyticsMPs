```python
df['EFG%'] = df['EFG%'].apply(lambda x: np.nan if x in ['-'] else x[:-1]).astype(float)/100
```

```python
df['ORB%'] = df['ORB%'].apply(lambda x: np.nan if x in ['-'] else x[:-1]).astype(float)/100
```

```python
df['TOV%'] = df['TOV%'].apply(lambda x: np.nan if x in ['-'] else x[:-1]).astype(float)/100
```

```python
df['FT RATE'] = df['FT RATE'].apply(lambda x: np.nan if x in ['-'] else x[:-1]).astype(float)/100
```

```python
df['O_EFG%'] = df['O_EFG%'].apply(lambda x: np.nan if x in ['-'] else x[:-1]).astype(float)/100
```

```python
df['O_ORB%'] = df['O_ORB%'].apply(lambda x: np.nan if x in ['-'] else x[:-1]).astype(float)/100
```

```python
df['O_TOV%'] = df['O_TOV%'].apply(lambda x: np.nan if x in ['-'] else x[:-1]).astype(float)/100
```

```python
df['O_FT RATE'] = df['O_FT RATE'].apply(lambda x: np.nan if x in ['-'] else x[:-1]).astype(float)/100
```

```python
year = df.loc[df['SEASON'] == '2020-21']
```

```python
hw9full = df[['SEASON', 'TEAM', 'EFG%', 'ORB%', 'TOV%', 'FT RATE', 'O_EFG%', 'O_ORB%', 'O_TOV%', 'O_FT RATE', 'WIN']].copy()
```

```python
hw9d = hw9full[['SEASON', 'TEAM', 'O_EFG%', 'O_ORB%', 'O_TOV%', 'O_FT RATE', 'WIN']].copy()
```

```python
hw9o = hw9full[['SEASON', 'TEAM','EFG%', 'ORB%', 'TOV%', 'FT RATE']].copy()
```

```python
hw9full
```

```python
hw9full['P1']  = hw9full['EFG%'] - hw9full['O_EFG%']
```

```python
hw9full['P2']  = hw9full['TOV%'] - hw9full['O_TOV%']
```

```python
hw9full['P3']  = hw9full['ORB%'] - hw9full['O_ORB%']
```

```python
hw9full['P4']  = hw9full['FT RATE'] - hw9full['O_FT RATE']
```

```python
hw9full.head(3)
```

```python
hw9full['SEASON'].value_counts()
```

```python
hw9train = hw9full.loc[(hw9full['SEASON'] != '2020-21')]
```

```python
hw9train = hw9train.drop(columns = ['EFG%','ORB%', 'TOV%', 'FT RATE', 'O_EFG%', 'O_ORB%', 'O_TOV%', 'O_FT RATE'])
```

```python
hw9train.head(2)
```

```python
hw9test = hw9full.loc[(hw9full['SEASON'] == '2020-21')]
```

```python
hw9test = hw9test[['SEASON','TEAM', 'P1', 'P2', 'P3', 'P4']]
```

```python
hw9test.head(2)
```

```python
X = hw9train[['P1', 'P2', 'P3', 'P4']]
y = hw9train['WIN']
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
lin_reg.score(X, y)
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
tX = hw9test[['P1', 'P2', 'P3', 'P4']]
```

```python
hw9test['Predicted Wins'] = lin_reg.predict(tX)
```

```python
hw9test
```

```python
sorted_final = hw9test.sort_values(by=['Predicted Wins'], ascending=False)
sorted_final
```
