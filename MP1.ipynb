**Building Runs Created Column**

```python
softball['RC'] = ((softball['H'] + softball['BB'] + softball['HBP']) * softball['TB'])/(softball['AB'] + softball['BB'] + softball['HBP'])
```

```python
softball['RC']
```

**Building Outs and CS Columns**

```python
softball.info()
```

```python
# This gets a little funky
```

```python
softball["SB-ATT"]= softball["SB-ATT"].str.split("-")
```

```python
softball['SB'] = [l[0] for l in softball['SB-ATT']]
```

```python
softball['ATT'] = [l[1] for l in softball['SB-ATT']]
```

```python
softball['SB'] =softball['SB'].fillna(0).astype(int)
```

```python
softball['ATT'] =softball['ATT'].fillna(0).astype(int)
```

```python
softball.drop('SB-ATT', axis=1, inplace=True)
```

```python
softball.head(1)
```

```python
softball['CS'] = softball['ATT']- softball['SB']
```

```python
softball.head(3)
```

```python
softball['Outs'] = (0.982*softball['AB']) - softball['H'] + softball['GDP'] + softball['SF'] + softball['SH'] + softball['CS']
```

```python
softball
```

```python
softball['Games'] = softball['Outs'] / 21
```

```python
softball['RC/G'] = softball['RC']/softball['Games']
```

```python
softball
```

```python
ordered_RCG = softball.sort_values(by = ['RC/G'], ascending = False)
```
