```python
candlestick = full[full["Year"] > 1957]
```

```python
candlestick
```

```python
candlestick = candlestick.drop(columns = ['G', 'GS', 'PA', 'R', 'H', '2B', '3B','RBI', 'SB', 'CS', 'BB', 'SO', 'BA', 'OBP', 'SLG', 'OPS', 'TB', 'GDP',
       'HBP', 'SH', 'SF', 'IBB', 'ROE', 'BAbip', 'tOPS+', 'sOPS+']) 
```

```python
candle_home = candlestick[candlestick['Split'] == 'Home']
```

```python
candle_home = candle_home.rename(columns={"Park_Factor_Splits": "Home_Park_Factor_Splits", "AB": "Home_ab", "HR": "Home_HR"})
```

```python
candle_away = candlestick[candlestick['Split'] == 'Away']
```

```python
def divide_two_cols_home(df):
    df['Home_Park_Factor_Splits'] = df['Home_HR'] / (df['Home_ab'])
    return df
```

```python
polo_home = polo_home.rename(columns={"Park_Factor_Splits": "Home_Park_Factor_Splits", "AB": "Home_ab", "HR": "Home_HR"})
```

```python
polo_home = polo_home.groupby(['Year']).apply(divide_two_cols_home) 
```

```python
polo_home
```

```python
def divide_two_cols_away(df):
    df['Away_Park_Factor_Splits'] = df['Away_HR'] / (df['Away_ab'])
    return df
```

```python
polo_away = polo_away.rename(columns={"Park_Factor_Splits": "Away_Park_Factor_Splits", "AB": "Away_ab", "HR": "Away_HR"})
```

```python
polo_away = polo_away.groupby(['Year']).apply(divide_two_cols_away) 
```

```python
polo_away
```

```python
polo_final = polo_home.merge(polo_away, how='inner', on='Year')
```

```python
polo_final
```

```python
polo_final['Annual_Park_Factor'] = polo_final['Home_Park_Factor_Splits'] / polo_final['Away_Park_Factor_Splits']
```

```python
polo_final
```

```python
polo_final.Home_HR.sum()
```

```python
polo_park_factor = ((polo_final.Home_HR.sum()) / (polo_final.Home_ab.sum())) / ((polo_final.Away_HR.sum()) / (polo_final.Away_ab.sum())) #This is way more simple
```

```python
polo_park_factor
```

```python
candlestick_park_factor = ((candle_home.Home_HR.sum()) / (candle_home.Home_ab.sum())) / ((candle_away.HR.sum()) / (candle_away.AB.sum()))
```
